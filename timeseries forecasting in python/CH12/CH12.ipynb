불필요한 컬럼 제거

```python
cols_to_drop = ['rain_1h', 'snow_1h']
df = df.drop(cols_to_drop, axis=1)

df.shape
```

```python
timestamp_s = pd.to_datetime(df['date_time']).map(datetime.datetime.timestamp)
```

- 날짜를 초로 표현

```python
This leads us to losing the cyclical nature of time, since 5pm is 5pm, no matter the day,
month, or year.
Therefore, we must apply a transformation to recuperate the cyclical behavior of time. A
simple way to do that is to apply a sine transformation. We know that the sine function is
cyclical, bounded between -1 and 1. Thus, this will help us regain part of the cyclical property
of time.fft = tf.signal.rfft(df['traffic_volume'])
f_per_dataset = np.arange(0, len(fft))

n_sample_h = len(df['traffic_volume'])
hours_per_week = 24 * 7
weeks_per_dataset = n_sample_h / hours_per_week

f_per_week = f_per_dataset / weeks_per_dataset

plt.step(f_per_week, np.abs(fft))
plt.xscale('log')
plt.xticks([1, 7], ['1/week', '1/day'])
plt.xlabel('Frequency')
plt.tight_layout()
plt.show()
```

초의 주기성을 고려하여 피처로 만든다.

```python
day = 24 * 60 * 60

df['day_sin'] = (np.sin(timestamp_s * (2*np.pi/day))).values
df['day_cos'] = (np.cos(timestamp_s * (2*np.pi/day))).values
```

```python
df = df.drop(['date_time'], axis=1)

df.head()
```

```python
n = len(df)

# Split 70:20:10 (train:validation:test)
train_df = df[0:int(n*0.7)]
val_df = df[int(n*0.7):int(n*0.9)]
test_df = df[int(n*0.9):]

train_df.shape, val_df.shape, test_df.shape
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaler.fit(train_df)

train_df[train_df.columns] = scaler.transform(train_df[train_df.columns])
val_df[val_df.columns] = scaler.transform(val_df[val_df.columns])
test_df[test_df.columns] = scaler.transform(test_df[test_df.columns])
```
